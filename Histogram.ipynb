```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import collections
```

```python
train1={'0': 4948, '1': 13861, '3': 8497, '2': 10585, '5': 6882, '4': 7458, '7': 5595, '6': 5727, '9': 4659, '8': 5045}
test1={'0': 1744, '1': 5099, '3': 2882, '2': 4149, '5': 2384, '4': 2523, '7': 2019, '6': 1977, '9': 1595, '8': 1660}
train = collections.OrderedDict(sorted(train1.items()))
test = collections.OrderedDict(sorted(test1.items()))
```

```python
X = np.arange(len(train))
plt.bar(X, train.values(), align='center', width=0.8)
plt.xticks(X, train.keys())
ymax = max(train.values()) + 1000
plt.ylim(0, ymax)
plt.suptitle('no of different labels in train data', fontsize=12)
plt.xlabel('labels', fontsize=10)
plt.ylabel('quantity of each labels', fontsize=10)
plt.show()
```

```python
X = np.arange(len(test))
plt.bar(X, test.values(), align='center', width=0.8)
plt.xticks(X, test.keys())
ymax = max(test.values()) + 1000
plt.ylim(0, ymax)
plt.suptitle('no of different labels in test data', fontsize=12)
plt.xlabel('labels', fontsize=10)
plt.ylabel('quantity of each labels', fontsize=10)
plt.show()
```
